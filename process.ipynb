```python
import pandas as pd
```

```python
time = pd.read_csv('Assignment Data.csv',usecols=[0]).dropna()
CPI = pd.read_csv('Assignment Data.csv',usecols=[1]).dropna()
wage = pd.read_csv('Assignment Data.csv',usecols=[2]).dropna()
equity_dividend_yield = pd.read_csv('Assignment Data.csv',usecols=[3]).dropna()
equity_price_index = pd.read_csv('Assignment Data.csv',usecols=[4]).dropna()
ten_year = pd.read_csv('Assignment Data.csv',usecols=[5]).dropna()
three_month = pd.read_csv('Assignment Data.csv',usecols=[6]).dropna()
```

```python
data=pd.concat([time,CPI,wage,equity_dividend_yield,equity_price_index,ten_year,three_month], axis=1, join='inner')
data = data.iloc[::-1]
data.columns = ["time","CPI","wage","equity index dividend yield","equity price index","10 year","3 month"]
data=data.set_index("time")
```

```python
data
```

```python
data.to_csv("raw data.csv")
```

```python
data = pd.concat([data["CPI"].pct_change(), data["wage"].pct_change(), data["equity index dividend yield"]+100*data["equity price index"].pct_change().dropna(),data["10 year"],data["3 month"]],axis=1,join='inner')
data.columns = ["CPI", "wage","return of equity","10 year","3 month"]
```
